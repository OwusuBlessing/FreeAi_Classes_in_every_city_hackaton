### Missing values

```python
df.isnull().sum()
```

```python
# Filling missin values in Year

median_value = df["Year"].median()

df["Year"].fillna(median_value,inplace = True)
df.isnull().sum()
```

```python
# fill missing values in distnance
# convert Distance and year columns to float
df["Distance"] = df["Distance"].astype(float)
df["Year"]  = df["Year"].astype(int)
```

```python
# Filling missing values in Type
mode = df["Type"].mode()
new_mode = mode[0]
```

```python
# fill nan values in Type

df["Type"] = df["Type"].fillna(new_mode)
df.head()
```

```python
# Group years based on average distance
group = pd.DataFrame(df.groupby("Year")['Distance'].mean())
type(group)
```

```python

group = group.reset_index(level=0)
group.head()
```

```python
# Fill nan value
median_value = df["Distance"].median()

group["Distance"].fillna(median_value,inplace = True)
```

```python
# convert year to list
year = list(group["Year"])
```

```python
# convert distance to list
Distance = list(group["Distance"])
```

```python
# Define function to replace distance based on avearage distance of each year
def fill_distance(cols):
    yea = cols[1]
    dist = cols[0]
    if pd.isnull(dist):
        if yea in year:
                 return Distance[year.index(yea)]
    else:
        return dist
df["Distance"] = df[["Distance","Year"]].apply(fill_distance,axis = 1)
df.head()
```

### Handling outliers in training data

```python
df.head()
```

```python
# Using mean imputation to handle outliers in distance

for i in df.iloc[:,7]:
    q1 =df.iloc[:,7].quantile(0.25)
    q3 = df.iloc[:,7].quantile(0.75)
    iqr = q3-q1
    lower_tail =  q1 - 1.5*iqr
    upper_tail = q3 + 1.5 * iqr
    if (i  > upper_tail) or (i < lower_tail):
        df.iloc[:,7] = df.iloc[:,7].replace(i,np.mean(df.iloc[:,7]))
    #else:
        #rain_data.iloc[:,5] = train_data["Distance"]
sns.boxplot(df["Distance"])
plt.title("Box plot mean imputation")
plt.show()
```

```python


for i in df["Year"]:
    q1 = df["Year"].quantile(0.25)
    q3 = df["Year"].quantile(0.75)
    iqr = q3-q1
    lower_tail =  q1 - 1.5*iqr
    upper_tail = q3 + 1.5 * iqr
    if (i  > upper_tail) or (i < lower_tail):
        df["Year"] = df["Year"].replace(i,np.mean(df["Year"]))
sns.boxplot(df["Year"])
plt.title("Box plot mean imputation")
plt.show()
```
